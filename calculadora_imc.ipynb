```python
nome = input('Digite seu nome')
peso = float(input('Digite seu peso em Kg e com "." para separar a casa decimal'))
altura = float(input('Digite sua altura em M e com "." para separar a casa decimal'))
IMC = peso/(altura**2)
print('Seu IMC é de',IMC)
```

```python
if IMC <= 17 :
  print(nome, 'seu IMC é de', IMC,'e você está muito abaixo do peso')

elif IMC <= 18.49 :
  print(nome, 'seu IMC é de', IMC, 'e você está abaixo do peso')

elif IMC <= 24.99 :
    print(nome, 'seu IMC é de', IMC, 'e você está com o peso normal')

elif IMC <= 29.99 :
    print(nome, 'seu IMC é de', IMC, 'e você está acima do peso')

elif IMC <= 34.99 :
    print(nome, 'seu IMC é de', IMC, 'e você está com grau de obesidade I')

elif IMC <= 39.99 :
    print(nome, 'seu IMC é de', IMC, 'e você está com grau de obesidade II (severa)')

else:
    print(nome, 'seu IMC é de', IMC, 'e você está com grau de obesidade III (mórbida)')

print('O IMC não deve ser considerado separadamente para um ato diagnostico. Procure um nutricionista ou um medico para melhores esclarecimentos ')

```
