###Como jogar
A Mega-Sena paga milh√µes para o acertador dos 6 n√∫meros sorteados. Ainda √© poss√≠vel ganhar pr√™mios ao acertar 4 ou 5 n√∫meros dentre os 60 dispon√≠veis no volante de apostas. Para realizar o sonho de ser o pr√≥ximo milion√°rio, voc√™ deve marcar de 6 a 15 n√∫meros do volante, podendo deixar que o sistema escolha os n√∫meros para voc√™ (Surpresinha) e/ou concorrer com a mesma aposta por 2, 4 ou 8 concursos consecutivos (Teimosinha).

###Sorteios
Os sorteios da Mega-Sena s√£o realizados duas vezes por semana, √†s quartas e aos s√°bados. Para adequar o n√∫mero do concurso da Mega da Virada, que deve ter final 0 ou 5 foram criadas as Mega-Semanas que s√£o exclusividade da Mega-Sena. Os sorteios ocorrem em datas predeterminadas ao longo do ano. Na ocasi√£o s√£o realizados tr√™s concursos semanais, √†s ter√ßas, quintas e s√°bados. Confira no cronograma de sorteio ou em Comunicados Importantes.

###Premia√ß√£o
O pr√™mio bruto corresponde a 43,35% da arrecada√ß√£o. Dessa porcentagem:

    35% s√£o distribu√≠dos entre os acertadores dos 6 n√∫meros sorteados (Sena);

    19% entre os acertadores de 5 n√∫meros (Quina);

    19% entre os acertadores de 4 n√∫meros (Quadra);

    22% ficam acumulados e s√£o distribu√≠dos aos acertadores dos 6 n√∫meros nos concursos de final 0 ou 5.

    5% ficam acumulados para a primeira faixa - sena - do √∫ltimo concurso do ano de final 0 ou 5 (Mega da Virada).


###Concurso Especial - Mega da Virada
O √∫ltimo concurso da Mega-Sena de final 0 ou 5 de cada ano civil, que tem a denomina√ß√£o comercial Mega da Virada, obedece √†s seguintes regras:

**Prazo de comercializa√ß√£o:**

Durante os meses de novembro e dezembro, com capta√ß√£o de apostas independente e concomitante com os demais concursos da modalidade, utilizando-se de volantes espec√≠ficos (a CAIXA informar√° com anteced√™ncia a data do in√≠cio das vendas e o n√∫mero do concurso da Mega da Virada, que ser√° sorteado no dia 31/12).

**Distribui√ß√£o do valor destinado ao pagamento dos pr√™mios:**

62% - primeira faixa - seis acertos (sena);

19% - segunda faixa - cinco acertos (quina);

19% - terceira faixa - quatro acertos (quadra).

**Composi√ß√£o da primeira faixa de premia√ß√£o (sena):**

- 62% do percentual destinado a pr√™mios, de acordo com a arrecada√ß√£o do respectivo concurso;

- O total acumulado para o √∫ltimo concurso de final zero ou cinco do ano civil;

- O total acumulado para o concurso de final zero ou cinco;

- O total acumulado na primeira faixa (sena) do concurso anterior, quando houver.

**Crit√©rio de acumula√ß√£o:**

- N√£o existindo apostas premiadas com seis n√∫meros (sena), o pr√™mio ser√° rateado entre os acertadores de cinco n√∫meros (quina);

- N√£o existindo apostas premiadas com seis e cinco n√∫meros, o pr√™mio ser√° rateado entre os acertadores de quatro n√∫meros (quadra);

- N√£o existindo apostas premiadas em quaisquer faixas de premia√ß√£o, os valores acumulam para o concurso seguinte, nas respectivas faixas.

```
repeticaoN = []
repeticao = {}
for n in listaNumeros:
  repeticaoN = (listaDezenas.count(n))
  repeticao[n]= n, repeticaoN
  

repeticaoDezena = pd.DataFrame.from_dict(repeticao, orient='index')
repeticaoDezena.columns = ['dezenas','repeti√ß√£o']

dezena = repeticaoDezena['dezenas']
repeticao = repeticaoDezena['repeti√ß√£o']

repeticao.sort_values(ascending=True).plot(kind='barh', 
                                           title='Dezenas por ordem de sorteio', 
                                           figsize=(10,25), 
                                           fontsize=14, 
                                           legend=True, 
                                           color='green',
                                           ylabel='Repeti√ß√£o',
                                           xlabel='Dezena',                                           
                                           )
```
