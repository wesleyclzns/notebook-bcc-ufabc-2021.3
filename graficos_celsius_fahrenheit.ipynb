Escreva os comandos para:

1. Criar um vetor de valores de graus Celsius de -100 a 100 

2. Usando a fórmula acima, calcule os valores em Fahrenheit para este mesmo vetor

3. Faça um gráfico mostrando a relação entre as temperaturas nas duas escalas

```python
#IMPOTAÇÃO DE BIBLIOTECAS

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
celsius = np.arange(-100,100.1,5)
fahrenheit = celsius * 1.8 + 32

plt.plot(celsius,  '.-', color='red', label='Celsius (°C)')
plt.plot(fahrenheit, '.-', color='green', label='Fahrenheit (°F)')

plt.grid(alpha=0.5)
plt.title(label='Relações entre graus Celsius e graus Fahrenheit')
plt.legend(title='Unidade de medida \n de temperatura')
```
